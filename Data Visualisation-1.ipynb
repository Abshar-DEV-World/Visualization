```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
# no.of products solds by person A and Person B , months in C
A = [22,24,23,24,23]
B = [21,22,23,24,25]
C = [1,2,3,4,5]
```

```python
plt.scatter(C,A) # c--> x axis, a--> Y axis
```

```python
plt.scatter(C,A, marker = 'D', color = 'Red')
plt.xlabel('Months')
plt.ylabel('Products Sold')
plt.title('Sales Performance of A')
plt.grid() # to enable background lines
plt.show() # it will hide unwanted information( <matplotlib.collections.PathCollection at 0x2119f569a50>)
```

```python
plt.scatter(C,A, marker = 'D', color = 'Red', label = 'Person A')
plt.scatter(C,B, marker = 'X', color = 'Blue', label = 'Person B')
plt.xlabel('Months')
plt.ylabel('Products Sold')
plt.title('Sales Performance of A')
plt.legend() # untill you enable legend values in written in label parameter are not displayed
plt.grid() # to enable background lines
plt.show() # it will hide unwanted information( <matplotlib.collections.PathCollection at 0x2119f569a50>)
```

```python
plt.subplot(121) # 1 row 2 columns 1st graph
plt.scatter(C,A, label = 'A')
plt.xlabel('Months')
plt.ylabel('Products Sold')
plt.title('Sales Performance of A')
plt.legend() # untill you enable legend values in written in label parameter are not displayed

plt.subplot(122) # 1 row 2 columns 2nd graph
plt.scatter(C,B, label = 'B')
plt.xlabel('Months')
plt.ylabel('Products Sold')
plt.title('Sales Performance of B')
plt.legend() # untill you enable legend values in written in label parameter are not displayed

plt.show()
```

```python
plt.subplot(221) # 2 row 2 columns 1st graph
plt.scatter(C,A, label = 'A')
plt.xlabel('Months')
plt.ylabel('Products Sold')
plt.title('Sales Performance of A')
plt.legend() # untill you enable legend values in written in label parameter are not displayed

plt.subplot(222) # 2 row 2 columns 2nd graph
plt.plot(C,A, label = 'A') # line plot
plt.xlabel('Months')
plt.ylabel('Products Sold')
plt.title('Sales Performance of A')
plt.legend() # untill you enable legend values in written in label parameter are not displayed

plt.subplot(223) # 2 row 2 columns 3rd graph
plt.scatter(C,B, label = 'B')
plt.xlabel('Months')
plt.ylabel('Products Sold')
plt.title('Sales Performance of B')
plt.legend() # untill you enable legend values in written in label parameter are not displayed


plt.subplot(224) # 1 row 2 columns 4th graph
plt.stackplot(C,B) # area plot
plt.xlabel('Months')
plt.ylabel('Products Sold')
plt.title('Sales Performance of B')

plt.subplots_adjust(wspace = 0.4, hspace= 0.5) # w --> width , h--> height
plt.show()
```

```python
fig, axs = plt.subplots(2,2)
# axs --> to control index values
# fig --> save the output generated by subplots
axs[0,0].scatter(C,A)
axs[0,1].plot(C,A)
axs[1,0].scatter(C,B)
axs[1,1].stackplot(C,B)

plt.show()
```

```python
# In statistics we have these two data types --> stats class
# continuous: mpg, displacement, Horse Power, weight, acceleration
# Discrete: cylinders, year, origin, name
```

#### Bar Plot

```python
# bar plot should be used for discrete data
# 2D 
# x axis : unique values 
# y axis : how many times each unique value appeared inside the data (frequency)
```

```python
df1[df1['origin'] == 1].shape
# 1 was appearing 248 times in df1
```

```python
df1[df1['origin'] == 2].shape
```

```python
df1[df1['origin'] == 3].shape
```

```python
uniq = [1,2,3]
freq = [248,70,79]
```

```python
plt.bar(uniq, freq)
plt.xlabel('Origin')
plt.ylabel('Frequency')
plt.title('Bar plot')
plt.xticks(uniq, [1,2,3])
plt.show()
```

```python
plt.bar(uniq, freq, tick_label = uniq)
plt.show()
```

#### Pie chart

```python
# discrete data 

plt.pie(freq, labels = uniq)
plt.show()
```

#### Box plot

```python
# boxplot is basically used to identify outliers in dataset --> stats class
# continuous data
```

```python
plt.boxplot(df1['weight']) # no outliers / no extreme values in weight column
plt.show()
```

```python
plt.boxplot(df1['acceleration']) # in acceleration we have outliers
plt.show()
```
